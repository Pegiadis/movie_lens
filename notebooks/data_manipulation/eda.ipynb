```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Histogram of 'vote_average'
plt.figure(figsize=(10, 6))
plt.hist(md['vote_average'], bins=30, color='skyblue')
plt.xlabel('Vote Average')
plt.ylabel('Frequency')
plt.title('Distribution of Vote Average')
plt.show()
```

```python
# Bar chart of 'original_language' (taking top 10 languages for simplicity)
plt.figure(figsize=(10, 6))
md['original_language'].value_counts()[:10].plot(kind='bar', color='skyblue')
plt.xlabel('Original Language')
plt.ylabel('Number of Movies')
plt.title('Number of Movies per Language')
plt.show()
```

```python
# Scatter plot of 'budget' vs 'revenue'
plt.figure(figsize=(20, 10))
plt.scatter(md['budget'], md['revenue'], color='skyblue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Budget vs Revenue')
plt.show()
```

```python
# Histogram of 'runtime'
plt.figure(figsize=(10, 6))
plt.hist(md['runtime'].dropna(), bins=30, color='skyblue')
plt.xlabel('Runtime (min)')
plt.ylabel('Frequency')
plt.title('Distribution of Movie Runtime')
plt.show()
```

```python
# Bar chart of 'genre' (taking top 10 genres for simplicity)
plt.figure(figsize=(10, 6))
md['genre'].value_counts()[:10].plot(kind='bar', color='skyblue')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Number of Movies per Genre')
plt.show()
```

```python
import seaborn as sns

# Heatmap of correlations
plt.figure(figsize=(10, 6))
sns.heatmap(md[['budget', 'popularity', 'revenue', 'runtime', 'vote_average', 'vote_count']].corr(), annot=True, fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Distribution of 'vote_average'
plt.figure(figsize=(10, 6))
sns.histplot(data=md, x='vote_average', kde=True)
plt.title('Distribution of Vote Average')
plt.show()
```

```python
data = pd.read_csv('../../data/kaggle_20m/movie_genres_ratings.csv')
```

```python
# Set the style of the visualization
sns.set(style="whitegrid")

# Create a histogram for the distribution of ratings
plt.figure(figsize=(10,6))
sns.histplot(data['rating'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(False)

# Show the plot
plt.show()
```

```python
# Summing each genre column to get the total count of movies per genre (since genres are binary)
genre_counts = data.iloc[:, 2:-1].sum().sort_values(ascending=False)

# Create a bar plot for the number of movies per genre
plt.figure(figsize=(12,8))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette="viridis")
plt.title('Number of Movies per Genre')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.grid(False)

# Show the plot
plt.show()
```

```python
# Calculate the average rating for each genre
genre_ratings = (data.iloc[:, 2:-1].multiply(data['rating'], axis="index").sum() / data.iloc[:, 2:-1].sum()).sort_values(ascending=False)

# Create a bar plot for the average rating per genre
plt.figure(figsize=(12,8))
sns.barplot(x=genre_ratings.values, y=genre_ratings.index, palette="coolwarm")
plt.title('Average Rating per Genre')
plt.xlabel('Average Rating')
plt.ylabel('Genre')
plt.grid(False)

# Show the plot
plt.show()
```
